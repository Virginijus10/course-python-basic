pirmiausiai sujunkite apatinę dalį, tada viršutinę *(yra persidengiančių stulpelių!)*, galiausiai viršų su apačia. Naudokite **.concat()** ir **.merge()** metodus. 

```python
top_left = os.path.join("tops", "top1-25-1.csv")
top_right = os.path.join("tops", "top1-25-2.csv")
bottom_left = os.path.join("tops", "top26-50-1.csv")
bottom_right = os.path.join("tops", "top26-50-2.csv")
```

```python
# Patikrinkite dabartinį darbinį aplankalą
print(os.getcwd())

# Patikrinkite, ar failai egzistuoja
print(os.path.exists(top_left))
print(os.path.exists(top_right))
print(os.path.exists(bottom_left))
print(os.path.exists(bottom_right))
```

```python
bottom_left_data = pd.read_csv(bottom_left)
bottom_right_data = pd.read_csv(bottom_right)
top_left_data = pd.read_csv(top_left)
top_right_data = pd.read_csv(top_right)

# Apjunkite duomenis iš apatinio kairiojo ir dešiniojo failų
data_bottom = pd.concat([bottom_left_data, bottom_right_data], axis=1)

# Apjunkite duomenis iš viršutinio kairiojo ir dešiniojo failų
data_top = pd.merge(top_left_data, top_right_data, on=["Track.Name", "Popularity"])

# Apjunkite apatinį ir viršutinį duomenų rinkinius
duomenys = pd.concat([data_bottom, data_top], sort=False)
print(duomenys)
```

# 5 Koks žanras pats populiariausias? Koks mažiausiai populiarus?
(pagal populiarumo vidurkį)

```python
populiariausi_z = pagal_zanra['Popularity'].mean().idxmax()
populiariausi_kiekis = pagal_zanra['Popularity'].mean().max()
print(populiariausi_z, populiariausi_kiekis)
maziausiai_populiarus_z = pagal_zanra['Popularity'].mean().idxmin()
maziausiai_populiarus_kieki = pagal_zanra['Popularity'].mean().min()
print(maziausiai_populiarus_z, maziausiai_populiarus_kieki)
```

# 6 Sukurkite lentelę, kurioje matytųsi, koks žanras turi aukščiausią vidurkį kiekviename indikatoriuje, bei pats vidurkis.
kad nereikėtų daug vargti su stulpelių pavadinimais, galima juos ištraukti su **.columns.tolist()**

```python
stul = duomenys.columns.tolist()[3:]
zanrai = pagal_zanra[stul].mean().idxmax()
skaiciai = pagal_zanra[stul].mean().max()
rezultatas = pd.DataFrame([stul, zanrai, skaiciai], ['indikatorius', 'zanrai', 'balai']).transpose()
```

# 8 Stulpelyje 'Genre' NaN reikšmes pakeiskite į 'pop'

```python
duomenys['Genre']= duomenys['Genre'].fillna('pop')
duomenys
```

# 9 Stulpelyje 'Popularity' trūkstamas reikšmes pakeiskite į stulpelio vidurkį


```python
duomenys['Popularity'] = duomenys['Popularity'].fillna(value=duomenys['Popularity'].mean())
```
